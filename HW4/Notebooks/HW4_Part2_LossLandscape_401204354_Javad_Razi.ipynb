```
%%time

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import numpy as np
import plotly.graph_objects as go
from tqdm import tqdm
import imageio
import os

# Load pre-trained model
model_name = 'gpt2'
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Set model to evaluation mode
model.eval()

# Define our input
input_text = "I have a dream"
inputs = tokenizer(input_text, return_tensors='pt')

# Compute the original loss
outputs = model(**inputs, labels=inputs["input_ids"])
original_loss = outputs.loss.item()

# Define two orthogonal random directions for each model parameter
direction1 = [torch.randn_like(p) for p in model.parameters()]
direction2 = [torch.randn_like(p) for p in model.parameters()]

# Normalize vectors
for p, d1, d2 in zip(model.parameters(), direction1, direction2):
    norm_p = torch.linalg.norm(p.flatten())
    d1.div_(torch.linalg.norm(d1.flatten())).mul_(norm_p)
    d2.div_(torch.linalg.norm(d2.flatten())).mul_(norm_p)

# Define the range to explore
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)

# Prepare to collect the losses
Z = np.zeros((x.size, y.size))

# Compute loss for each direction
for i in tqdm(range(x.size), desc="x progress"):
    for j in tqdm(range(y.size), desc="y progress", leave=False):
        # Perturb the model parameters
        for p, d1, d2 in zip(model.parameters(), direction1, direction2):
            p.data.add_(x[i]*d1 + y[j]*d2)

        # Compute the loss
        outputs = model(**inputs, labels=inputs["input_ids"])
        Z[i, j] = outputs.loss.item()

        # Revert the model parameters
        for p, d1, d2 in zip(model.parameters(), direction1, direction2):
            p.data.sub_(x[i]*d1 + y[j]*d2)
```

```
# Check whether direction1 and direction2 are really orthogonal, using cosine similarity
import torch.nn.functional as F

# Concatenate all tensors in direction1 and direction2
d1_concat = torch.cat([t.flatten() for t in direction1])
d2_concat = torch.cat([t.flatten() for t in direction2])

# Compute cosine similarity
cos_sim = F.cosine_similarity(d1_concat.unsqueeze(0), d2_concat.unsqueeze(0))
print(f"Cosine similarity: {cos_sim.item()}")  # Should be close to 0 if orthogonal
```

```
# Create 3D plot using the meshgrid
import numpy as np
import plotly.graph_objects as go

fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y)])
fig.update_layout(title='Loss Landscape', autosize=False,
                  width=700, height=700,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()
```

## Interpretation of the Results

The cosine similarity value obtained from the analysis is approximately 0.00027, which is very close to 0. This indicates that the two random directions generated for the exploration of the loss landscape are nearly orthogonal, as expected. Orthogonality in this context implies that the changes in the model parameters along these two directions are almost independent of each other. This is a desirable property as it ensures that the exploration of the loss landscape is being conducted in two distinct directions, providing a more comprehensive view of the landscape.

The 3D surface plot visualizes the loss landscape of the GPT-2 model. The X and Y axes represent the perturbations in the two random directions, and the Z axis represents the loss. The plot's wavy structure and color gradient from purple (lowest points, value 20) to yellow (highest points, value 160) represent variations in the loss values. The landscape appears to be relatively smooth with some noticeable peaks and valleys, indicating areas of parameter space where the model's performance is sensitive to changes in its parameters.

The orthogonality of the random directions ensures a thorough exploration of the landscape, and the 3D plot provides a visual representation of the model's performance sensitivity to changes in its parameters. The smoothness of the landscape suggests that the model is well-optimized, but the presence of peaks and valleys indicates areas where the model's performance could potentially be improved. This analysis forms a crucial part of understanding and interpreting the performance of large language models. It highlights the importance of visualizing and understanding the high-dimensional loss landscapes of such models, which can guide future research and development efforts in the field of natural language processing.
